```python
nominal = ['Loan_ID', 'Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status']

ordinal = ['Dependents', 'Credit_History', 'Loan_Amount_Term']

continuous = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']

discrete = ['Dependents', 'Credit_History', 'Loan_Amount_Term', 'Loan_Status']
```

```python
print("Nominal Columns:", nominal)
print("Ordinal Columns:", ordinal)
print("Continuous Columns:", continuous)
print("Discrete Columns:", discrete)
```

```python
print(data.dtypes)
```

```python
data['Gender'].fillna(data['Gender'].mode()[0], inplace=True)
```

```python
data['Married'].fillna(data['Married'].mode()[0], inplace=True)
```

```python
data['ApplicantIncome'] = data['ApplicantIncome'].fillna(data['ApplicantIncome'].mean())
data['CoapplicantIncome'] = data['CoapplicantIncome'].fillna(data['CoapplicantIncome'].mean())
data['LoanAmount'] = data['LoanAmount'].fillna(data['LoanAmount'].mean())
###- continuous numerical use mean
```

```python
data['Loan_Amount_Term'] = data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].median())
data['Credit_History'] = data['Credit_History'].fillna(data['Credit_History'].median())
###- discrete numrical use median 
```
