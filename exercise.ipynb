## Exercise dataset

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("exercise")
```

```python
print("Shape:", df.shape)
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
print(df.head(3))
```

```python
df.describe()
```

```python
avg_pulse = df.groupby('kind')['pulse'].mean()
print("\nAverage Pulse by Exercise Kind:\n", avg_pulse)
```

```python
plt.figure(figsize=(8,5))
plt.bar(avg_pulse.index, avg_pulse.values, color='green')
plt.title("Average Pulse by Exercise (Music Proxy)")
plt.ylabel("Pulse")
plt.show()
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='kind', y='pulse', data=df, palette='pastel')
plt.title("Pulse Distribution by Exercise Type")
plt.ylabel("Pulse")
plt.show()
```
